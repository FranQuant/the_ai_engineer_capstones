We define a tiny policy that decides whether to call a calculator tool or respond directly. The loop demonstrates observe → plan → act → learn.

```python
from __future__ import annotations # import modules

def parse(text: str): # define function parse()
    # Parse simple patterns like 'add 2 and 3' or 'divide 9 over 3'.
    t = text.lower().strip() # set t
    if t.startswith('add'): # conditional branch
        import re # import modules
        m = re.search(r'(\d+)\D+(\d+)', t) # set m
        if m: # conditional branch
            return ('add', int(m.group(1)), int(m.group(2))) # conditional branch
 # return value
    if 'divide' in t: # conditional branch
        import re # import modules
        m = re.search(r'(\d+)\D+(\d+)', t) # set m
        if m and int(m.group(2))!=0: # conditional branch
            return ('div', int(m.group(1)), int(m.group(2))) # conditional branch
 # return value
    return ('echo', text) # return value

def calculator(
    a: int,
    b: int,
    op: str,
) -> int | float:
    # define function calculator()
    # Execute a tiny arithmetic step with a guard.
    if op=='add': # conditional branch
        return a + b # conditional branch
 # return value
    if op=='div': # conditional branch
        return a / b if b!=0 else float('inf') # conditional branch
 # return value
    raise ValueError('unknown op') # step

def policy(obs: str) -> str: # define function policy()
    # Choose between a tool call and a direct echo.
    p = parse(obs) # set p
    if p[0]=='add': # conditional branch
        return f"{p[1]} + {p[2]} = {calculator(p[1], p[2], 'add')}"
 # return value
    if p[0]=='div': # conditional branch
        return f"{p[1]} / {p[2]} = {calculator(p[1], p[2], 'div')}"
 # return value
    return 'echo: ' + str(p[1]) # return value

for q in ['add 2 and 3','divide 9 over 3','hello agent']: # loop over items
    print(policy(q)) # print output
```
