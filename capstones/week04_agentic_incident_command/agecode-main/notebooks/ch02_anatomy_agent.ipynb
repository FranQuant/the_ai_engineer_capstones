We define a small observation, a policy that decides when to call a tool, and a minimal in-notebook memory list.

```python
from __future__ import annotations # import modules

memory: list[str] = []  # episodic log

def tool_calculator(a:int,b:int,op:str)->str: # define function tool_calculator()
    if op=='add': # conditional branch
        return str(a+b) # conditional branch
 # return value
    if op=='mul': # conditional branch
        return str(a*b) # conditional branch
 # return value
    return 'error: unknown op' # return value

def policy(observation:str)->str: # define function policy()
    if observation.startswith('add'): # conditional branch
        return tool_calculator(2,3,'add') # return value
    return 'echo' # return value

observation = 'add two numbers' # set observation
out = policy(observation) # set out
memory.append('obs='+observation+' out='+out) # call memory.append()
print(out) # print output
print(memory[-1]) # print output
```
