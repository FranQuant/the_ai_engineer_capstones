Parse short notes for dates and amounts, then compose a one-line summary.

```python
import re, json # import modules
def extract_fields(text:str): # define function extract_fields()
    d=re.search(r'\b(\d{4}-\d{2}-\d{2})\b', text) # set d
    a=re.search(r'\b(\d+(?:\.\d+)?)(?:\s*(USD|EUR))?\b', text) # set a
    return {
        'date': d.group(1) if d else 'n/a',
        'amount': a.group(1) if a else 'n/a',
        'currency': (a.group(2) if a and a.group(2) else 'n/a'),
    }  # return value
def build_summary(doc_id:str, f:dict): # define function build_summary()
    parts=['doc='+doc_id, 'date='+f.get('date','n/a')] # set parts
    amount = f.get('amount', 'n/a')
    currency = f.get('currency', 'n/a')
    if amount != 'n/a':
        fragment = f"amount={amount} {currency}"
    else:
        fragment = 'no amounts found'
    parts.append(fragment)  # call parts.append()
    s=', '.join(parts) # set s
    return s if len(s)<=120 else s[:117]+'…' # return value
doc='On 2024-09-01 revenue rose to 3.5 USD.' # set doc
fields=extract_fields(doc) # set fields
print(json.dumps({'doc_id':'ex1','extracted':fields})) # print output
print(build_summary('ex1',fields)) # print output
```
