A tiny research → signal → report pipeline with deterministic rules.

```python
POS={'beats','surge','record'} # set POS
NEG={'misses','downgrade','cut'} # set NEG
def extract(text:str): # define function extract()
    t=text.lower() # set t
    if any(w in t for w in POS): # conditional branch
        return {'cue':'positive', 'reason':'positive phrasing'} # conditional branch
 # return value
    if any(w in t for w in NEG): # conditional branch
        return {'cue':'negative', 'reason':'negative phrasing'} # conditional branch
 # return value
    return {'cue':'neutral','reason':'no strong cue'} # return value
def decide(ticker:str,cue:str): # define function decide()
    allow={'ACME','FOO','BAR'} # set allow
    if ticker not in allow: # conditional branch
        return {
            'status': 'error',
            'signal': 'n/a',
            'reason': 'unknown ticker',
        }  # conditional branch
 # return value
    sig='up' if cue=='positive' else 'down' if cue=='negative' else 'flat' # step
    return {'status':'ok','signal': sig} # return value
def report(ticker,dec,ex): # define function report()
    signal = dec.get('signal', '')
    reason = ex.get('reason', '')
    return f"{ticker}: signal={signal} because {reason}"  # return value
ex=extract('ACME beats estimates') # set ex
dec=decide('ACME',ex['cue']) # set dec
print(report('ACME',dec,ex)) # print output
```
